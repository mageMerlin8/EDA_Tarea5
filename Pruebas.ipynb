```python
def creaArbol(nums):
    miarbol = ArbolB()
    for i in nums:
        miarbol.insert(i)
    return miarbol

def eliminaDeArbol(arb, nums):
    for i in nums:
        arb.delete(i)
    return arb

def eliminaYMuestraDeArbol(arb, nums):
    print('Arbol original:')
    arb.draw()
    for i in nums:
        print('Se intent√≥ borrar ' + str(i) + ' :')
        arb.delete(i)
        arb.draw()
    return arb
    
```

```python
listaArbol = [70, 50, 100, 30, 60, 80, 115, 20, 40, 65, 75, 87, 112, 35, 63, 68, 67]
listaElim = [70, 80, 100, 50]
miarbol = creaArbol(listaArbol)
miarbol.draw()
eliminaYMuestraDeArbol(miarbol, listaElim)
```
