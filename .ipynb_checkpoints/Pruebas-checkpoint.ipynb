## Se crearon las siguientes funciones para la prueba de lo arboles binarios de búsqueda
## creaArbol(nums):
### params:
nums: una lista de números a agregar al arbol (se agregan en ese orden)  
regresa: un arbol binario con los datos de nums  
## eliminaYMuestraDeArbol(arb, nums):
Esta función intenta eliminar todos los elementos de nums del arbol en orden y muestra el arbol despues de cada operación
### params:
arb: arbol  
nums: lista de números a eliminar


```python
def creaArbol(nums):
    miarbol = ArbolB()
    for i in nums:
        miarbol.insert(i)
    return miarbol

def eliminaDeArbol(arb, nums):
    for i in nums:
        arb.delete(i)
    return arb

def eliminaYMuestraDeArbol(arb, nums):
    print('Arbol original:')
    arb.draw()
    for i in nums:
        print('Se intentó borrar ' + str(i) + ' :')
        arb.delete(i)
        arb.draw()
    return arb
    
```

Se crea un arbol con una lista de números y se muestra:  

```python
listaArbol = [70, 50, 100, 30, 60, 80, 115, 20, 40, 65, 75, 87, 112, 35, 63, 68, 67]
miarbol = creaArbol(listaArbol)
miarbol.draw()
```

Después se define la lista de números a eliminar y se ejecuta la función para eliminarlos y mostrarlos:

```python
listaElim = [70, 80, 100, 50]
eliminaYMuestraDeArbol(miarbol, listaElim)
```
